```python
year = "2014"
```

```python
def find_performance_data_box(boxes):
    for box in boxes:
        if "performance data" in box.text.lower() and "competition" in box.text.lower():
            return box
```

```python
def get_player_data(player_link, year):
    try:
        browser.get(player_link.replace("profil", "leistungsdaten")+"/plus/0?saison=" + str(int(year) - 2))
        boxes = browser.find_elements_by_class_name("box")
        box = find_performance_data_box(boxes)
        height = browser.find_element_by_css_selector('span[itemprop="height"]').text.replace(",","").replace(" m","")
        game_stats = box.find_element_by_tag_name("tfoot").find_elements_by_class_name("zentriert")
        game_stats = list(map(lambda x: x.text, game_stats))
        minutes = box.find_element_by_tag_name("tfoot").find_elements_by_class_name("rechts")[-1]
        minutes = minutes.text.replace(".","").replace("'","")
        game_stats.append(minutes)
        game_stats.append(height)
        game_stats.append(player_link)
        return game_stats
    except:
        return ['-','-','-','-','-','-','-','-',player_link]
```

```python
world_cup_years = ["1978"]
for year in [world_cup_years[0]]:
    print(year)
    scrape_df = pd.read_csv("data/to_scrape/" + year + ".csv")
    player_data = scrape_df.link.apply(lambda x: get_player_data(x, year))
    player_data = player_data.apply(lambda x:pd.Series(x))
    player_data.columns = ["appearances","goals","assists","yellow_cards","yellow_red_cards","red_cards","minutes","height","link"]
    #write scraped data
    final_player_data = pd.merge(scrape_df, player_data, "left", on="link")
    player_data.to_csv("data/to_scrape/scraped_" + year + ".csv", index=False)
```

```python
year = "2018"
```

```python
def get_players_from_team_string(team_string):
    remove_digits = str.maketrans('', '', digits)
    return list(map(lambda x: x.translate(remove_digits).replace("  ","").replace("--","").replace("\n","")[1:], team_string.split("_")))
```

```python
df_2018 = pd.read_csv("2018.csv")
```

```python
players_2018 = list(set(get_players_from_team_string(df_2018.team1_lineup.values.sum() + df_2018.team1_lineup.values.sum())))
```

```python
players_2018
```

```python
players_df = pd.read_csv("data/teams/2018.csv")
players_df = players_df.drop("Unnamed: 0", axis=1)
```

```python
players_df = players_df[players_df.name.isin(players_2018)]
```

```python
player_data = players_df.link.apply(lambda x: get_player_data(x, year))
```

```python
player_data = player_data.apply(lambda x:pd.Series(x))
player_data.columns = ["appearances","goals","assists","yellow_cards","yellow_red_cards","red_cards","minutes","height","link"]
#write scraped data
final_player_data = pd.merge(players_df, player_data, "left", on="link")
```

```python
final_player_data.to_csv("data/to_scrape/scraped_" + year + ".csv", index=False)
```
